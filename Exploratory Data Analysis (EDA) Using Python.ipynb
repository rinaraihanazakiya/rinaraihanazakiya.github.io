```python
#for uploading file
from google.colab import files

#for reading file
import io

#for importing file
import pandas as pd

#for visualization
import seaborn as sns

uploaded= files.upload()
df=pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
```

```python
#show columns
df.columns
```

```python
df.head()
```

```python
#show data info and check for null value
df.info()
```

```python
# convert total charges to float
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors="coerce")
```

```python
df.isnull().sum()
```

```python
#missing value handling
df[df['TotalCharges'].isna()]
#drop rows with missing value
df=df.dropna()
df.isnull().sum()
```

```python
#drop irrelevant columns
df=df.drop(columns=['customerID'])
```

```python
#convert seniorcitizen into category
df['SeniorCitizen']=df['SeniorCitizen'].astype('object')
df.dtypes
```

```python
#show statistics summary
df.describe()
```

```python
#visualization
import seaborn as sns

sns.countplot(data=df,x="Churn")
```

Berdasarkan grafik, terlihat bahwa lebih banyak pelanggan yang tidak churn (No Churn) dibandingkan pelanggan yang Churn.

```python
sns.countplot(data=df,x="gender",hue="Churn")
```

Berdasarkan grafik, terlihat bahwa gender tidak berpengaruh terhadap Churn.

```python
sns.boxplot(data=df,x="tenure",y="Churn")
```

Berdasarkan grafik, terlihat bahwa pelanggan yang berusia kurang dari 30 tahun didominasi mengalami churn. Sedangkan yang tidak mengalami churn (no churn) sampai berumur 60 tahun. Bagi pelanggan yang berusia 70 tahun, sangat sedikit yang mengalami churn.

```python
sns.boxplot(data=df,x="MonthlyCharges",y="Churn")
```

Berdasarkan grafik, terlihat bahwa pelanggan dengan biaya bulanan berada diantara rentang 60-100 cenderung mengalami churn, sedangkan pelanggan yang biaya bulanannya lebih kecil, mendominasi tidak churn.

```python
sns.boxplot(data=df,x="TotalCharges",y="Churn")
```

Terlihat bahwa pelanggan yang memiliki total pembayaran mencapai 2000, mereka mengalami churn karena memutuskan untuk tidak membayar lagi.

```python
sns.countplot(data=df,x="Contract",hue="Churn")
```

```python
#get dummies to transform data
df2=pd.get_dummies(df,drop_first=True)
```

```python
#correlation

# 0 no correlation
# 1 positive correlation
# -1 negative correlation

df3=df2.corr()['Churn_Yes'].sort_values().reset_index()

df3
```

```python
#drop Churn_Yes
df3=df3[~df3['index'].str.contains('Churn_Yes')]
```

```python
chart = sns.barplot(data=df3,x="index",y="Churn_Yes")

for item in chart.get_xticklabels():
  item.set_rotation(90)
```
